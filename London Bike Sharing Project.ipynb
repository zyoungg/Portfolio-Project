```python
# Create column dictionary according to Kaggle metadata
column_dict = {
    'timestamp' : 'time',
    'cnt' : 'bike_count',
    't1' : 'real_temp_C',
    't2' : 'feels_like_temp_C',
    'hum' : 'humidity_percent',
    'wind_speed' : 'wind_speed_kmh',
    'weather_code' : 'weather'
}
# Rename columns
df.rename(columns = column_dict, inplace=True)
```

```python
# Create dictionaries for season and weather according to Kaggle metadata
season_dict = {
    '0.0' : 'Spring',
    '1.0' : 'Summer',
    '2.0' : 'Fall',
    '3.0' : 'Winter'
}
weather_dict = {
    '1.0' : 'Clear',
    '2.0' : 'Scattered clouds',
    '3.0' : 'Broken clouds',
    '4.0' : 'Cloudy',
    '7.0' : 'Rain',
    '10.0' : 'Rain with thunderstorm',
    '26.0' : 'Snowfall',
    '94.0' : 'Freezing fog'
}
```

```python
# Mapping corresponding values from dictionary to 'season' and 'weather' columns 
df['season'] = df['season'].astype(str).map(season_dict)
df['weather'] = df['weather'].astype(str).map(weather_dict)
```
